# ch 10. File Handling in Python
- Python too supports file handling and allows users to handle files i.e., to read and write files, along with many other file handling options, to operate on files.
- In Python, a file operation takes place in the following order:
 - Open a file
 - Read or write (perform operation)
 - Close the file


## Opening Files in Python
f = open(filename, mode)

- w: open an existing file for a write operation. If the file already contains some data then it will be overridden but if the file is not present then it creates the file as well.
- a:  open an existing file for append operation. It wonâ€™t override existing data.

## 2.write() mode

```python
file=open('write_test.txt', 'w')
file.write("This is the write mode. ")
file.write("It allows us to write in a particular file. ")
file.close()
```

### 3. append() mode


```python
file=open('write_test.txt', 'a')
file.write("Append mode won't override existing data. ")
file.close()
```

### 4. Move File Pointer: seek
The position (index) of the first character in files is zero, just like the string index.

```python
file=open('test1.txt', 'r')
file.seek(11)
print(file.read())
file.close()
```

 ## search for a string in text file

```python
def search_string(file_path, word):
    file=open(file_path, 'r')
    content=file.read()
    if word in content:
        print("The word does exist in the file.")
    else:
        print("The word does not exist in the file.")
```

```python
search_string('test1.txt', 'Python')
```

```python
search_string('test1.txt', 'python')
```

```python
search_string('test1.txt', 'basic')
```

## Search file for a string and Print its line and line number

```python
def search_word(file_path, word):
    file=open(file_path, 'r')
    lines=file.readlines()
    for line in lines:
        if line.find(word)!=-1:
            print(word, ' The word does exist in the file')
            print('Line number ', line.index(line))
            print('Line ', line)
            
search_word('test1.txt', 'Python')
```
